# 13.find the average attack for lenegndary and non - legendary pokemons

```python
pokemon.head(2)
```

```python
average_attack=pd.DataFrame(pokemon.groupby('legendary')['attack'].mean())
average_attack
```

# 14.find the average defense for lenegndary and non - legendary pokemons

```python
pokemon.head(2)
```

```python
average_defence=pd.DataFrame(pokemon.groupby('legendary')['defense'].mean())
average_defence
```

# 15.compare the average defense and attack for legendary and non-legendry pokemons

```python
average_defence_attack=pd.merge(average_attack,average_defence,how='inner',on='legendary')
average_defence_attack
```
